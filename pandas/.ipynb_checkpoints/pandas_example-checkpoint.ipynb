```python
import pandas as pd
```

```python
# Load data from a CSV file
df = pd.read_csv("netflix_titles.csv")
```

```python
# show first 2 rows
print(df.head(2))
# show last 2 rows
print(df.tail(2))
```

```python
# show all the columns names
print(df.columns)
```

```python
# show first 5 rows of a specific column
print(df["title"][0:5])
```

```python
# show first 5 rows of multiple columns
print(df[["title", "director"]][0:5])
```

```python
# select rows based on a condition
df.loc[df["director"] == "Peter Jackson"]
```

```python
df.describe()
```

```python
# ascending order, smallest to largest
df.sort_values("release_year")[0:5]
```

```python
# decending order, largest to smallest
df.sort_values("release_year", ascending=False)[0:5]
```

```python
df.sort_values(["release_year", "date_added"], ascending=False)[0:5]
```
